````python
## Solution copied from my submission from last year for the course Programming in Python for Text / 2020

def read_csv(input_file, delimiter=","):
    '''
    TODO DOC
    '''
    list_of_dicts = []
    with open(input_file, "r") as infile:
        lines = infile.read().split('\n')
    headers = lines[0]
    for line in lines[1:]:
        line_list = line.split(delimiter)
        header_list = headers.split(delimiter)
        row_dict = dict(zip(header_list,line_list))
        list_of_dicts.append(row_dict)
    return list_of_dicts

# test your function here
filename = "../../Data/csv_data/trump_facebook.tsv"
status_updates = read_csv(filename, delimiter="\t") 
status_updates[:2]
````

### 1b. Find the status updates with the most responses

Define a function called **`get_update_most_responded_to()`** that has the following parameters: 
* **`status_updates`** (positional parameter) 
* **`response_type`** (keyword parameter with default string `"likes"`) 

The fuction should find the status update that received the highest number of possible reactions to a Facebook status (emoji such as 'angrys', 'comments', 'hahas', etc. - anything that starts with 'num_'). It should return three strings: the **`status_message`**, the **`status_type`** and the **`status_link`** of this particular status update.


````python
## Solution copied from my submission from last year for the course Programming in Python for Text / 2020

def get_update_most_responded_to(status_updates, response_type="likes"):
    '''
    TODO DOC
    '''
    response = "num_" + response_type.lower()
    all_values = [int(x[response]) for x in status_updates]
    max_number_reactions = max(all_values)
    for i in status_updates:
        if int(i[response]) == max_number_reactions:
            message = i['status_message']
            status_type = i['status_type']
            link = i['status_link']
    return message,status_type,link#,max_number_reactions

#test function            
get_update_most_responded_to(status_updates,'wows')
````

### 1c. Find the longest status updates

Define a function called **`get_longest_update()`** that has the following parameters: 
* **`status_updates`** (positional parameter) 
* **`length_type`** (keyword parameter with default string `"tokens"`). 

The function should find the longest update. By default, the fuction should find the status update that is the longest in terms of number of tokens. Also implement the options to find the longst status update in terms of characters or sentences in the message. These options should be carried out when `length_type` is changed to `"sentences"` or `"characters"` 

The function should return the status message (called `'status_message'` in the json) of the longest update as a string. 

**Attention**: It is recommended to use NLTK for this exercise. 


````python
## Solution copied from my submission from last year for the course Programming in Python for Text / 2020

def get_longest_update(status_updates, length_type = 'tokens'):
    from nltk.tokenize import word_tokenize
    from nltk.tokenize import sent_tokenize
    ## tokens (words)
    if length_type == 'tokens':
        token_len = []
        for i in status_updates:
            tokens = word_tokenize(i['status_message'])
            amount_of_tokens = len(tokens)
            token_len.append(amount_of_tokens)
        max_tokens = max(token_len)
        for i in status_updates:
            if len(word_tokenize(i['status_message'])) == max_tokens:
                status_message = i['status_message']
    ## sentences
    elif length_type == 'sentences':
        sent_len = []
        for i in status_updates:
            tokens = sent_tokenize(i['status_message'])
            amount_of_tokens = len(tokens)
            sent_len.append(amount_of_tokens)
        max_tokens = max(sent_len)
        for i in status_updates:
            if len(sent_tokenize(i['status_message'])) == max_tokens:
                status_message = i['status_message']
    ## characters
    elif length_type == 'characters':
        char_len = []
        for i in status_updates:
            len_characters = len(i['status_message'])
            char_len.append(len_characters)
        max_characters = max(char_len)
        for i in status_updates:
            if len(i['status_message']) == max_characters:
                status_message = i['status_message']
     
    return status_message
    
        
# test function
len(get_longest_update(status_updates,'characters'))
````

The function should return **`filtered_status_updates`**, which is a list of dictioaries with all information about the status updates (same format as the input argument `'status_updates'`). 

**Attention**: It is highly recommended to use NLTK for this exercise. Make sure that you **tokenize** the messages before you look for keywords. 

````python
## Solution copied from my submission from last year for the course Programming in Python for Text / 2020

def get_update_with_keywords(status_updates,keywords,case_sensitive=False):
    from nltk.tokenize import word_tokenize
    updates_with_keywords = []
    if case_sensitive == False:
        for i in status_updates:
            for j in keywords:
                if j in i['status_message'].lower():
                    updates_with_keywords.append(i)    
    if case_sensitive == True: 
        for i in status_updates:
            for j in keywords:
                if j in i['status_message']:
                    updates_with_keywords.append(i)
    return updates_with_keywords
                

# test with these keywords; also experiment with other keywords
keywords = ["clinton", "obama"]

# test your function here
get_update_with_keywords(status_updates,keywords,True) == get_update_with_keywords(status_updates,keywords,False)
````

````python
## Solution copied from my submission from last year for the course Programming in Python for Text / 2020

def get_laureates(dict_prizes,year=None,category=None):
    names_nobel = []
    if year != None:
        if category != None:
            for i in dict_prizes['prizes']:
                if i['year'] == str(year) and i['category'] == category:
                    for i in i['laureates']:
                        names_nobel.append(i['firstname'] + ' ' + i['surname'])
        else: 
            for i in dict_prizes['prizes']:
                if i['year'] == str(year):
                    for i in i['laureates']:
                        names_nobel.append(i['firstname'] + ' ' + i['surname'])
    elif year == None and category != None:
        for i in dict_prizes['prizes']:
                if i['category'] == category:
                    for i in i['laureates']:
                        names_nobel.append(i['firstname'] + ' ' + i['surname'])
                        
    else:
        for i in dict_prizes['prizes']:
            for i in i['laureates']:
                names_nobel.append(i['firstname'] + ' ' + i['surname'])
    return names_nobel
                    


year = 2018
category = "peace"
# test your function here

get_laureates(dict_prizes, year, category),get_laureates(dict_prizes,category="physics")
````
